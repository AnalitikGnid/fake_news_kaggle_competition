```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
data = pd.read_csv('encoded_text_Tfidf.csv')
```

```python
data[data != 0] = 1
```

```python
all_values = []
for column in tqdm(data.columns):
    values = data[column].sum(axis=0)
    all_values.append(values)

all_values
```

```python
plt.figure(figsize=(20, 10))
sns.histplot(all_values, bins=100)
plt.xlabel('Number of documents containing the word')
plt.ylabel('Number of words')
plt.title('Distribution of words by number of documents containing them')
plt.show()
```

```python
all_values = pd.Series(all_values)
```

```python
all_values.sort_values(ascending=False, inplace=True)
```
